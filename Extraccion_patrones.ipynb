```
#prueba del método
fpgrowth(df.values, supp=20) #extraer todos los itemsets frecuentes de df con al menos 20% de ocurrencias
```

##Remove NaN from transactions

```
data = []
for i in range(df.shape[0]): #para todas las filas
  fila = df.iloc[i,:].values #se recupera la fila con todas sus columnas
  fila = fila[~pd.isnull(fila)] #mantener todos los elementos que no sean null/nan
  data.append(fila.tolist()) #agregar fila a la data
data

#df.fillna('', inplace=True) #versión corta
```

```
#extraer todos los itemsets con 20% de freq mínima, y tamaño mínimo de 2
r = fpgrowth(data, supp=20, zmin=2)
df_ = pd.DataFrame(r)
df_.columns = ['Itemset','Freq']
df_.sort_values(by='Freq',ascending=False,inplace=True)
df_
```

```
#extraer los itemsets con 20 como mínimo de ocurrencias (en valor absoluto) y máximo 3 items
fpgrowth(data, supp=-20, zmax=3)
```

#Real Data: [LastFM](https://www.last.fm/es/)

```
lastfm = pd.read_csv("https://www.biz.uiowa.edu/faculty/jledolter/DataMining/lastfm.csv")
lastfm.shape
```

```
lastfm.head(20)
```

##Data Pre-processing

```
#Remove columns
lastfm = lastfm[['user','artist']]
```

```
#Drop duplicates
lastfm = lastfm.drop_duplicates()
lastfm.shape
```

```
lastfm.head()
```

```
#Count how many artists
print(len(lastfm.artist.unique()))
print(lastfm.artist.unique())
```

```
#Count how many users
print(len(lastfm.user.unique()))
print(lastfm.user.unique())
```

##Transforming the data into a transactional dataset

```
trans = []
for u in lastfm.user.unique(): #para cada usuario recuperar su lista de artistas favoritos
  trans.append(list(lastfm[lastfm.user == u]['artist'].values))
  #print(lastfm[lastfm.user == u]['artist'])
print(len(trans))
```

```
print('El usuario',lastfm.user.unique()[0],'escucha los artistas',trans[0])
print('El usuario',lastfm.user.unique()[100],'escucha los artistas',trans[100])
print('El usuario',lastfm.user.unique()[1000],'escucha los artistas',trans[1000])
print('El usuario',lastfm.user.unique()[5000],'escucha los artistas',trans[5000])
print('El usuario',lastfm.user.unique()[10000],'escucha los artistas',trans[10000])
```

##Mining itemsets

```
fpgrowth(trans, supp=1, zmin=2)
```

##Mining Association Rules

###support 1% and confidence 40%

```
ar = fpgrowth(trans, target='r', supp=1, conf=40, report='aCl')
ar

# conf(('coldplay', 'the beatles') -> 'radiohead') = freq('coldplay', 'the beatles','radiohead')/freq('coldplay', 'the beatles')
#=293/665=0.4406
```

```
df_ar = pd.DataFrame(ar)
df_ar.columns = ['Consequent','Antecedent','Freq','Conf','Lift']
df_ar = df_ar[df_ar.Lift >= 1]
df_ar.sort_values(by='Conf',ascending=False,inplace=True)
df_ar#.head(100)
```
